```python
import firebase_admin
from firebase_admin import credentials
from firebase_admin import firestore
import pandas as pd
```

```python
cred = credentials.Certificate("serviceAccountKey.json")
firebase_admin.initialize_app(cred)
```

```python
db = firestore.client()
```

```python
dataset = pd.read_csv('data\RFM_Scores.csv')
dataset.head()
```

```python
customerID_list = dataset['CustomerID'].astype(str).tolist()
RFMSegment_list = dataset['RFMSegment'].tolist()
customerID_list
```

```python

# Tuple'lar kullanarak yeni bir szlk oluturma
data = dict(zip(customerID_list, RFMSegment_list))
data.values()
data.keys()
```

```python
doc_ref = db.collection("MyData").document("Segment")
doc_ref.set(data)
```
